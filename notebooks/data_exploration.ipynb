# FridgeWise-AI — Data Exploration

Quick overview of processed datasets used by the recommender pipeline.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
processed = ROOT / 'data/processed'

recipes = pd.read_csv(processed / 'clean_recipes.csv')
interactions = pd.read_csv(processed / 'clean_interactions_train.csv')
fridge = pd.read_csv(processed / 'user_fridge_inventory.csv')
products = pd.read_csv(processed / 'clean_open_food_products.csv')

print(f'Recipes: {len(recipes):,}')
print(f'Train interactions: {len(interactions):,}')
print(f'Fridge items: {len(fridge):,} ({fridge.user_id.nunique()} users)')
print(f'Open Food Facts products: {len(products):,}')
```

```python
fridge.groupby('storage_type').size().plot(kind='bar', title='Fridge items by storage', color='#2ecc71')
```

```python
fridge['days_to_expiry'].hist(bins=15, color='#3498db')
import matplotlib.pyplot as plt
plt.title('Days to expiry (synthetic fridges)')
plt.xlabel('Days')
plt.show()
```
